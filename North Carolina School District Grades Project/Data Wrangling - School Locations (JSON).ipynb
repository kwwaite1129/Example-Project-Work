```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = 'https://opendata.arcgis.com/datasets/dea6ff0e8b4743a0ba361e13a85a4c70_3.geojson'
r = requests.get(url)

if 'json' in r.headers.get('Content-Type'):
    js = r.json()
else:
    print('Response content is not in JSON format.')
    js = 'spam'
```

```python
data_items = js['features']
data = []
for item in data_items:
    data.append(item['properties'])
data
```

```python
df = pd.DataFrame(data)
df
```

```python
# Print out a list of all of the columns names to analyze and see which ones to keep
df.columns
```

```python
# First Transformation: Keep only the columns needed for analysis. Reorder the columns to make more readable
df = df[['OBJECTID', 'LEA_SCHOOL','SCHOOL_NAM','PHYS_ADDR','PHYS_CITY','PHYS_ZIP','PHONE','URL_ADDRES','COUNTY',
         'ACCR_STAT', 'SCH_DESG', 'SCH_TYPE', 'SCH_PTYPE', 'SCH_CTYPE']]
df
```

```python
# Identify which columns have NaN's and investigate why
df.isna().sum()
```

```python
# Second Transformation
# Odd - I know there are values with Na's, yet none are showing up in counts. They must be in there as blank space.
# Replace all blank spaces with NA's
df = df.replace(r'^\s*$', np.nan, regex=True)
df.isna().sum()
```

```python
# Lucky for me, my dataset is pretty clean and does not contain missing data for any of the major columns
list_of_cols = list(df.columns)
# create a function that iterates over a list and adds data and their counts to a dictionary
def getDuplicatesWithCount(listOfElems):
    ''' Get frequency count of duplicate elements in the given list '''
    dictOfElems = dict()
    # Iterate over each element in list
    for elem in listOfElems:
        # If element exists in dict then increment its value else add it in dict
        if elem in dictOfElems:
            dictOfElems[elem] += 1
        else:
            dictOfElems[elem] = 1    
 
    # Filter key-value pairs in dictionary. Keep pairs whose value is greater than 1 
    # i.e. only duplicate elements from list.
    dictOfElems = { key:value for key, value in dictOfElems.items() if value > 1}
    # Returns a dict of duplicate elements and thier frequency count
    return dictOfElems

# Create a function that says if there are not duplicates in a column or identifies how many duplicates and what
# word is duplicated how many times
def printDuplicates(col_name):
    empty_dict = {} 
    for col in list_of_cols:
        dictOfElems = getDuplicatesWithCount(col_name)
    
    if dictOfElems == empty_dict:
        print("No Duplicates")
    else:
        for key, value in dictOfElems.items():
                print(key , ' :: ', value)
```

```python
# Identify if there are any duplicates in my primary column, LEA_SCHOOL, which is equal to school ID
printDuplicates(df['LEA_SCHOOL'])
```

```python
# There is one set of duplicates, let's investigate
duplicate_df = df[df['LEA_SCHOOL'] == '92P000']
duplicate_df
```

```python
# Third Transformation:
# It looks like we have a duplicate in the data set. I will go ahead and drop one of these rows
df.drop(df[df['OBJECTID'] == 2574].index, inplace = True)
printDuplicates(df['LEA_SCHOOL'])
```

```python
# I want to see if there are any duplicate school addresses
printDuplicates(df['PHYS_ADDR'])
```

```python
# There appear to be addresses that show up more than once. Let's investigate
values = df['PHYS_ADDR'].value_counts().keys().tolist()
counts = df['PHYS_ADDR'].value_counts().tolist()
res = {values[i]: counts[i] for i in range(len(values))}
newList = []
for (key, value) in res.items():
    if value > 1:
        newList.append(key)
dup_add = df[df['PHYS_ADDR'].isin(newList)]
dup_add = dup_add.sort_values(by=['PHYS_ADDR'])
dup_add
```

```python
# Fourth Transformation: Clean Data
# Most of the schools that have duplicate addresses are co-op type schools that have multiple programs at one
# location. However, two schools have the incorrect addresses. I will adjust
# School 650355 has an incorrect address. Will update
df.at[2381, 'PHYS_ADDR'] = '3702 Princess Place Drive'
# School 340436 has an incorrect address. Will update
df.at[952, 'PHYS_ADDR'] = '1394 Thurmond Street'
```

```python
# Review the data types for each column to ensure the data is in the correct format
df.dtypes
```

```python
# Fifth Transformation
# All the data types look ok except for the phone number. It is stored as a number. Did some research and standards
# say to store phone number as a string. Convert PHONE col from an int64 to a string
df = df.astype({'PHONE': 'str'})
df.dtypes
```

```python
df['PHONE'].value_counts()
```

```python
# It looks like the data in PHONE for unknown numbers is 0. I want to change that to Unknown for consistency
df.loc[df.PHONE == '0', 'PHONE'] = 'Unknown'
df['PHONE'].value_counts()
```
